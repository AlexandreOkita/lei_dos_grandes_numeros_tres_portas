```python
import random as rd
import pandas as pd
```

```python
def simular_jogo(resposta):
    premio = rd.randint(0,2)
    escolha = rd.randint(0,2)
    if premio == escolha:
        porta_escolhida = 1 #positivo representa True
    else:
        porta_escolhida = -1 #negativo representa False
    if resposta == 'sim':
        porta_escolhida *= -1 
    if porta_escolhida > 0:
        return 1
    else:
        return 0
```

```python
lista_probs_caso = []
for i in range(1, 1000):
    num_vitorias = 0
    for num_testes in range(i):
        num_vitorias += simular_jogo('sim')
    prob = num_vitorias/i
    lista_probs_caso.append(prob)
```

```python
probs_numTestes =  pd.DataFrame({'Probabilidade de vencer com sim': lista_probs_caso})
probs_numTestes.plot.line()
```

```python
lista_probs_caso_nao = []
for i in range(1, 1000):
    num_vitorias = 0
    for num_testes in range(i):
        num_vitorias += simular_jogo('nao')
    prob = num_vitorias/i
    lista_probs_caso_nao.append(prob)
```

```python
probs_numTestes_nao =  pd.DataFrame({'Probabilidade de vencer com nao': lista_probs_caso_nao})
probs_numTestes_nao.plot.line()
```
